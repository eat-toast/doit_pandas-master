# 4-2 matplotlib 자유자대로 사용하기

```
# tips에는 팁에 관련된 데이터가 들어있다.
# 지불금액, 팁, 성별, 흡연유무, 요일, 시간, 진체인원
tips = sns.load_dataset('tips')
print(tips.head())
print(type(tips))
```

### Hist

```
# 기본틀(fig)를 마련하고 그래프 격자(axes1)을 구성
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)

# hist 메소드에 특정 열을 전달하면 hist가 완성
axes1.hist(tips['total_bill'], bins = 10)
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Freq')
axes1.set_ylabel('Total Bill')
```

### # 산점도 그리기

```
# 산점도 그리기
scatter_plot = plt.figure() # 기본틀 만들기
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot og Total Bill vs Tip')
axes1.set_xlabel('Total BIll')
axes1.set_ylabel('Tip')
```

### 박스 플랏

```
# 박스 플랏
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([ tips[tips['sex'] == 'Female']['tip'],
                tips[tips['sex'] == 'Male']['tip']],
              labels = ['Female', 'Male'] )

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot og Tips by Sex')
```

### apply 활용

```
def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
```

```
# apply에 함수를 넣는것은 R과 동일한데 사용하는 구조가 다르네.. 
# Series.apply(함수)
tips['sex_color'] = tips['sex'].apply(recode_sex) 
```

```
# 산점도를 그릴 때, 점의 크기도 지정할 수 있다.
# 이번에는 테이블당 인원수도 점의 크기로 그려본다.
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
  x = tips['total_bill'],
  y = tips['tip'],
  s = tips['size']*20,
  c = tips['sex_color'],
  alpha = 0.5)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

# 4-3 seaborn 자유자재로 사용하기

```
# matplot보다 깔끔한 그래프를 그릴 수 있다. 
# R에서의 ggplot같은 느낌스트

import seaborn as sns
tips = sns.load_dataset('tips')
```

```
# ax에 자꾸 대입하는 이유는 무엇일까
# ax = plt.subplot()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

### 산점도 그리기

```
ax = sns.regplot(x = 'total_bill', y='tip', data = tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```
ax = sns.regplot(x = 'total_bill', y='tip', data = tips, fit_reg = False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

### 산점도와 히스트 한번에!! 

```
joint = sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize = 10, y = 1.03)
```

```
joint = sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize = 10, y = 1.03)
```

### 2차원 밀집도 그리기

```
ax = sns.kdeplot(data = tips['total_bill'],
                  data2 = tips['tip'] ) # shade = True
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('tip')
```

### 관계 그래프 그리기

```
fig = sns.pairplot(tips)
```

```
# pairplot은 위 아래로 같은 그림이다. 
# 이를 옵션으로 변경할 수 있다.
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug = True)
```
